```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
from google.colab import drive
drive.mount('/content/drive')
path = '/content/drive/MyDrive/Data Mining/jobsheet5/Restaurant.csv'
df_res = pd.read_csv(path, header=None)
```

```python
dataset_res = df_res.apply(lambda x: [str(i).strip() for i in x], axis=1).tolist()
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset_res).transform(dataset_res)
df_transformed = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# menggunakan min_support 0.1 karena data cukup besar (179rb baris)
frequent_itemsets = apriori(df_transformed, min_support=0.1, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```

```python
hasil_res = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
print("Hasil Association Rules: Restaurant.csv ")
display(hasil_res)
```

```python
pat = '/content/drive/MyDrive/Data Mining/jobsheet5/Chapter05DataSet.csv'
df_chap = pd.read_csv(pat, header=None)
```

```python
# ubah setiap sel menjadi format 'NamaKolom=Nilai'
# contoh: Gender jadi 'Gender=M', Working jadi 'Working=No'
dataset_chap = []
for index, row in df_chap.iterrows():
    transaction = [f"{col}={row[col]}" for col in df_chap.columns]
    dataset_chap.append(transaction)
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset_chap).transform(dataset_chap)
df_transformed = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# menggunakan min_support 0.3 agar mendapatkan pola yang cukup kuat
frequent_itemsets = apriori(df_transformed, min_support=0.3, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
```

```python
hasil_chap = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
print("Hasil Association Rules: Chapter05DataSet.csv")
display(hasil_chap)
```
