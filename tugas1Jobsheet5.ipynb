```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
from google.colab import drive
drive.mount('/content/drive')
path = '/content/drive/MyDrive/Data Mining/jobsheet5/GroceryStoreDataSet.csv'
df_raw = pd.read_csv(path, header=None)
```

```python
dataset = df_raw[0].str.split(',').tolist()
```

```python
te = TransactionEncoder()
te_data = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_data, columns=te.columns_)
```

```python
frequent_items = apriori(df, min_support=0.1, use_colnames=True)
frequent_items
```

```python
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.5)
rules
```

```python
hasil = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
print("Hasil Association Rules:")
display(hasil)
```
